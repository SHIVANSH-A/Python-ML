```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.arange(1,30)
y=np.arange(31,60)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Scatter Plot')
plt.savefig("test.png")
```

```python
y=x*x
```

```python
plt.plot(x,y,'b*', linestyle='dashed', linewidth=2, markersize=10)
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('2D Plot')
plt.savefig("test.png")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b*', linestyle='dashed', linewidth=2, markersize=10)
plt.subplot(2,2,2)
plt.plot(x,y,'r*', linestyle='dashed', linewidth=2, markersize=10)
plt.subplot(2,2,3)
plt.scatter(x,y,c='b')
plt.subplot(2,2,4)
plt.scatter(x,y,c='g')
```

```python
x=np.arange(1,11)
y=5*x+5
```

```python
x=np.arange(0,4*np.pi, 0.2)
y=np.sin(x)
plt.plot(x,y,'b')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Sine Wave Plot')
plt.savefig("test.png")
```

```python
x=np.arange(0,4*np.pi, 0.2)
y=np.cos(x)
plt.plot(x,y,'b')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Cosine Wave Plot')
plt.savefig("test.png")
```
